```python
from training.run import Run
from evaluation import metrics, plots
```

```python
def make_config(model_name, pca):
    if pca != 0:
        return {'model_name': model_name, 'features': 'all', 'train_size': 100, 
                  'test_size': 100, 'pca': pca}
    else:
        return {'model_name': model_name, 'features': ['mass', 'd2'], 'train_size': 100, 
                  'test_size': 100}
```

```python
def test(config):
    run = Run(config)
    metrics.print_metrics(run)
    plots.plot_roc_curve(run)
    plots.plot_scores(run)
```

## Scikit-learn SVM

```python
config = make_config('sklearn_svm', pca=0)
pca_config = make_config('sklearn_svm', pca=2)
```

### No PCA

```python
test(config)
```

### PCA

```python
test(pca_config)
```

```python
config = make_config('svm_classical', pca=0)
pca_config = make_config('svm_classical', pca=2)
```
